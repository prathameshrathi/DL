```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tensorflow.keras import datasets,models,optimizers,layers
import tensorflow as tf
from sklearn.preprocessing import LabelBinarizer
```

```python
df = pd.read_csv('C:/Users/hp/PICT/DL/mnist_train/mnist_train.csv')
df
```

```python
X_train = df.iloc[0:,1:]
# X_train.shape
print(X_train)
```

```python
Y_train = df.iloc[0:,0:1]
print(Y_train)
```

```python
df2 = pd.read_csv("C:/Users/hp/PICT/DL/mnist_test/mnist_test.csv")
df2
```

```python
X_test = df2.iloc[0:,1:]
Y_test = df2.iloc[0:,0:1]
print(X_test)
print(Y_test)
```

```python
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
X_train = np.array(X_train).reshape(60000,28,28,1)
X_train.shape
```

```python
X_test = np.array(X_test).reshape(10000,28,28,1)
```

```python
lb = LabelBinarizer()
Y_train = lb.fit_transform(Y_train)
Y_test = lb.transform(Y_test)
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28,28,1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))
```

```python
model.summary()
```

```python
#model compilation
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
epochs = 3
h = model.fit(X_train, Y_train, epochs=epochs, validation_data=(X_test, Y_test))
```

```python
plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0,epochs),h.history["loss"],label="train_loss")
plt.plot(np.arange(0,epochs),h.history["val_loss"],label="val_loss")
plt.plot(np.arange(0,epochs),h.history["accuracy"],label="accu")
plt.plot(np.arange(0,epochs),h.history["val_accuracy"],label="val_accu")
plt.title("Loss/accuracy")
plt.show()
```
